```python
import torch
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('train.csv', header=0)
```

```python
df = df[:100]
```

```python
plt.scatter(df.x, df.y)
```

```python
w = 1.0
def forward(x):
    return x * w

def loss(x, y):
    y_pred = forward(w)
    ans = (y_pred - y ) * (y_pred - y )
    return ans
```

```python
w_list = []
mse_list = []
SIZE = df.size
for w in np.arange(0.0, 10.1, 0.1):
    print(f"w: {w}")
    l_sum = 0
    for x_val, y_val in zip(df.x, df.y):
        y_pred_val = forward(x_val)
        l = loss(x_val, y_val)
        l_sum += l
#         print(f"\t {x_val}, {y_val}, {y_pred_val}, {l}")
    
    print(f"MSE: {l_sum/SIZE}")
    w_list.append(w)
    mse_list.append(l_sum/SIZE)
plt.plot(w_list, mse_list)
plt.ylabel('Loss')
plt.xlabel('w')
plt.show()
```
